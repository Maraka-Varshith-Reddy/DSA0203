```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("image_3b8a5190.png")

if img is None:
    print("Error: Image not found or could not be loaded. Please ensure 'image_3b8a5190.png' is in the correct directory.")
else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 100, 200)
    cv2.imwrite('Canny_Edges_3rd.jpg', edges)

    cv2_imshow(edges)
```
