```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("image_3b8a5190.png")

if image is None:
    print("image is not avilable")
else:
    blur = cv2.GaussianBlur(image, (23, 21), 0)

    cv2.imwrite("blurred_image.jpg", blur)

    print("Blurred Image")
    cv2_imshow(blur)
```
