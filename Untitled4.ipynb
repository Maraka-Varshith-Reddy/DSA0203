```python
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread("gods.jpg")

if img is None:
    print("Error: Image 'gods.jpg' not found. Please upload the image or provide the correct path.")
else:
    rotated_img=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)

    rvsrotated_img=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)

    cv2_imshow(rotated_img)
    cv2_imshow(rvsrotated_img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
